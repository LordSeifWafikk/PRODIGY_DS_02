```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

#### Data Cleaning

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Fill missing 'Survived' with mode since it's categorical and should be either 0 or 1
df['Survived'].fillna(df['Survived'].mode()[0], inplace=True)

# Fill missing 'Gender' with mode
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)

# Fill missing 'Fare' with median
df['Fare'].fillna(df['Fare'].median(), inplace=True)

# Display the cleaned dataset
print(df.isnull().sum())
```

We could also clean the data by dropping the empty categories ".dropna" 

#### Analysis

```python
# Age distribution
sns.histplot(data=df, x='Age', kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
# Gender distribution
sns.countplot(data=df, x='Gender')
plt.title('Gender Distribution')
plt.show()
```

```python
# Survived vs. Not Survived by Gender
sns.countplot(data=df, x='Survived', hue='Gender')
plt.title('Survived vs. Not Survived by Gender')
plt.show()
```

```python
# Age distribution by Survived
sns.histplot(data=df, x='Age', hue='Survived', kde=True)
plt.title('Age Distribution by Survived')
plt.show()
```

```python
# Fare distribution by Survived
sns.histplot(data=df, x='Fare', hue='Survived', kde=True)
plt.title('Fare Distribution by Survived')
plt.show()
```

```python
# Pairplot of numerical variables
sns.pairplot(df, hue='Survived', vars=['Age', 'Fare', 'Pclass'])
plt.title('Pairplot of Numerical Variables by Survived')
plt.show()
```

```python
# Select only the numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Correlation matrix
corr = numeric_df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Age vs. Fare by Survived
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived')
plt.title('Age vs. Fare by Survived')
plt.show()
```

```python
# Age vs. Fare by Pclass
sns.scatterplot(data=df, x='Age', y='Fare', hue='Pclass' ,palette= 'muted')
plt.title('Age vs. Fare by Pclass')
plt.show()
```

```python
#Exclude non-numeric columns
numeric_df = df.select_dtypes(include=[np.number])
sns.pairplot(numeric_df, diag_kind='kde')
plt.show()
```
